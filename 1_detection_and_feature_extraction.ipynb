# Detection and Feature Extraction

```python
import numpy as np
import pandas as pd
from time import process_time
from maad.rois import find_rois_cwt
from maad import sound
import settings
from soundclim_utilities import (features_to_csv, 
                                 batch_find_rois, 
                                 batch_feature_rois, 
                                 listdir_pattern)
```

```python
def feature_extraction(quality, id_rois):
    settings_MAAD['opt_shape_str'] = quality    
    features = batch_feature_rois(rois_list, settings_MAAD, settings.path_audio['train'])    
    df_features = features_to_csv(features).loc[id_rois, :]
    df_features.reset_index(inplace=True, drop=True)
    return df_features
```

```python
def feature_extraction_process_time(quality):
    settings_MAAD['opt_shape_str'] = quality 
    t_start = process_time()
    for i in range(loops_number):
        features = batch_feature_rois(rois_list, settings_MAAD, settings.path_audio['train'])  
    t_stop = process_time() 
    return (t_stop-t_start)/loops_number
```

```python
path_id_rois = './data/id_rois.csv'           #path to file with id of regions of interest annotated
path_low  = './results/df_features_low.csv'   #path to save file with low resolution
path_med  = './results/df_features_med.csv'   #path to save file with med resolution
path_high = './results/df_features_high.csv'  #path to save file with high resolution
path_times = './results/df_times.csv'         #path to save file with process times
```

# Detection of Regions of Interest

```python
settings_MAAD = settings.features                           
flist = listdir_pattern(settings.path_audio['train'], ends_with='.wav') 
flist = pd.DataFrame(flist, columns=['fname'])                          
detection_data = batch_find_rois(flist, settings.detection,         
                                     settings.path_audio['train'])      
rois_list = detection_data['detections']
```

# Feature Extraction

```python
#get the list of id of the regions of interest annotated 
#to filter from all the regions of interest detected
df_id_rois = pd.read_csv(path_id_rois)
id_rois = df_id_rois['id_rois'].to_list()
```

```python
df_features_low = feature_extraction('low', id_rois)
```

```python
df_features_med = feature_extraction('med', id_rois)
```

```python
df_features_high = feature_extraction('high', id_rois)
```

```python
#save ROI's features 
df_features_low.to_csv(path_low, index=False)
df_features_med.to_csv(path_med, index=False)
df_features_high.to_csv(path_high, index=False)
```
